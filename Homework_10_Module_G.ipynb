Chapter 21.4: Simpson’s Rule

```python
import numpy as np

a = 0
b = np.pi
n = 13
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = np.cos(x)

I_simp = (h/3) * (f[0] + 3*sum(f[:n-2:2]) \
            + 4*sum(f[1:n-1:2]) + f[n-1])
err_simp = 4 - I_simp

print(I_simp)
print(err_simp)
```

Chapter 21.5: Computing Integrals in Python- f(x)=cos(x) from 0 to π with a discretization step of 0.05

```python
from scipy.integrate import cumtrapz
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('seaborn-poster')

x = np.arange(0, np.pi, 0.05)
F_exact = -np.sin(x)
F_approx = cumtrapz(np.cos(x), x)

plt.figure(figsize = (10,6))
plt.plot(x, F_exact)
plt.plot(x[1::], F_approx)
plt.grid()
plt.tight_layout()
plt.title('$F(x) = \int_0^{x} cos(y) dy$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['Exact with Offset', 'Approx'])
plt.show()
```
