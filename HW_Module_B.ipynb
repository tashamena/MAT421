17.2 Linear Interpolation at x=4.4 based on the data x = [4, 1, 9], y = [5, 2, 3]. 

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

x = [4, 1, 9]
y = [5, 2, 3]
f = interp1d(x, y)
y_hat = f(4.4)
print('ŷ=',y_hat)

plt.figure(figsize = (10,8))
plt.plot(x, y, '-ob')
plt.plot(1.5, y_hat, 'ro')
plt.title('Linear Interpolation at x = 7.2')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

17.2 Cubic Spline Interpolation for the data set  x = [0, 1, 2] and y = [2, 7, 2] for 0≤x≤2.

```python
from scipy.interpolate import CubicSpline
import numpy as np


x = [0, 1, 2]
y = [2, 7, 2]

f = CubicSpline(x, y, bc_type = 'natural')
x_new = np.linspace(0, 2, 100)
y_new = f(x_new)

plt.figure(figsize = (10,8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
